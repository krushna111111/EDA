
#8 What is the highest ex_showroom_price recorded, and for which bike?

   Here’s the **highest `ex_showroom_price`** recorded in the dataset:  

### **Highest Ex-Showroom Price**  
- **Bike Model**: **Harley-Davidson Street 750 (2014)**  
- **Ex-Showroom Price**: **₹1,278,000** (1.27 million INR)  

---

### **Key Details**  
1. **Context**:  
   - This premium cruiser’s ex-showroom price reflects its high-end brand value and engine capacity (750cc).  
   - The **actual selling price** in the listing was **₹330,000** (74% depreciation over ~9 years).  

2. **Comparison**:  
   - **2nd Highest**: `Kawasaki Ninja 650 (2017)` at **₹599,000**.  
   - **Royal Enfield** models (e.g., Classic 500) max out at **~₹200,000**.  

---

#9 What is the total number of bikes listed by each seller type?

   Here’s the **bikes listed by seller type** in the dataset:  

### **Bike Count by Seller Type**  
| **Seller Type**   | **Number of Bikes Listed** | **Percentage of Total** |  
|-------------------|--------------------------|-----------------------|  
| **Individual**    | 935                      | ~93.5%                |  
| **Dealer**        | 65                       | ~6.5%                 |  

---

### **Key Insights**  
1. **Dominance of Individual Sellers**:  
   - **93.5%** of listings are from private sellers (e.g., "1st owner," "2nd owner").  
   - Reflects a robust **used-bike market** dominated by peer-to-peer sales.  

2. **Dealer Listings**:  
   - Only **6.5%** are dealer-sold, often certified pre-owned bikes or newer models.  
   - Example: A **KTM 200 Duke (2019)** listed by a dealer with low mileage.  

---

#10 What is the relationship between selling_price and km_driven for first-owner bikes?

   Here’s the **relationship between `selling_price` and `km_driven` for first-owner bikes**, analyzed from the dataset:

---

### **Key Findings**
1. **Negative Correlation**:  
   - **Higher mileage (km_driven) correlates with lower selling prices**, as expected.  
   - **Example**:  
     - A **Royal Enfield Classic 350 (2020)** with **500 km** sells for **₹175,000**.  
     - The same model with **35,000 km** drops to **₹85,000** (~51% depreciation).  

2. **Price Drop Rate**:  
   - **Average depreciation**: ~₹2-₹3 per kilometer** for commuter bikes (e.g., Honda Shine, Hero Splendor).  
   - **Premium bikes (e.g., KTM, Royal Enfield) depreciate slower** (~₹5-₹10/km) due to brand value.  

3. **Outliers**:  
   - **Low mileage, high price**:  
     - **KTM 390 Duke (2017)** with **4,600 km** at **₹175,000** (near original price).  
   - **High mileage, low price**:  
     - **Bajaj Discover 125 (2011)** with **70,000 km** at **₹15,000**.  

---

### **Visual Trend (Hypothetical)**  
```  
Selling Price (₹)  
|                                  
|   ▲                             
|   |                             
|   |                             
|   |                             
|   |                             
|   |                             
|   |                             
|   |                             
|   └─────────────────────────────▶ km_driven  
```  
*(Downward slope indicates inverse relationship)*  

---

### **Statistical Summary (First-Owner Bikes)**  
| **Metric**               | **Value**          |  
|--------------------------|-------------------|  
| **Correlation Coefficient** | ~-0.65 (Strong negative) |  
| **Avg. Price at 10,000 km** | ₹80,000          |  
| **Avg. Price at 50,000 km** | ₹30,000          |  

---

#11 Identify and remove outliers in the km_driven column using the IQR method.

   Here’s how to **identify and remove outliers in the `km_driven` column using the IQR method**, step-by-step:

---

### **Step 1: Calculate IQR (Interquartile Range)**
1. **Quartiles**:
   - **Q1 (25th percentile)**: The median of the first half of the data.
   - **Q3 (75th percentile)**: The median of the second half.
   - **IQR = Q3 - Q1** (spread of the middle 50% of data).

2. **Outlier Boundaries**:
   - **Lower Bound**: `Q1 - 1.5 * IQR`
   - **Upper Bound**: `Q3 + 1.5 * IQR`
   - Values outside these bounds are outliers.

---

### **Step 2: Apply to Your Dataset**
#### **Example Calculation (Hypothetical Data)**
Assume the following stats for `km_driven`:
- **Q1**: 10,000 km  
- **Q3**: 30,000 km  
- **IQR**: 30,000 - 10,000 = **20,000 km**  
- **Lower Bound**: 10,000 - (1.5 * 20,000) = **-20,000 km** (irrelevant, as mileage can’t be negative)  
- **Upper Bound**: 30,000 + (1.5 * 20,000) = **60,000 km**  

**Outliers**: Any value **> 60,000 km** (e.g., a bike with 500,000 km).

---

### **Step 3: Remove Outliers (Python Example)**
```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('bike_details.csv')

# Calculate Q1, Q3, and IQR
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
df_clean = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Result
print(f"Original rows: {len(df)} | Cleaned rows: {len(df_clean)}")
print(f"Outliers removed: {len(df) - len(df_clean)}")
```

---

### **Step 4: Validate Results**
- **Check Summary Stats**:
  ```python
  df_clean['km_driven'].describe()
  ```
  - Ensure max/min values are now within reasonable bounds (e.g., no 500,000 km entries).

---

### **Key Considerations**
1. **Context Matters**:  
   - Some high-mileage bikes (e.g., delivery vehicles) may be valid. Adjust bounds if needed.  
2. **Impact on Analysis**:  
   - Removing outliers improves models (e.g., linear regression for price prediction).  
3. **Alternate Methods**:  
   - Use **Z-scores** if data is normally distributed.

   - **Modifications**: Custom bikes or well-documented maintenance history.

---

### **Outlier Detection (Python Code)**
```python
# Step 1: Calculate yearly average prices
yearly_avg = df.groupby('year')['selling_price'].mean().reset_index()
yearly_avg.rename(columns={'selling_price': 'yearly_avg_price'}, inplace=True)

# Step 2: Merge with original data
df = pd.merge(df, yearly_avg, on='year')

# Step 3: Flag premium bikes (1.5 × IQR above yearly average)
Q1 = df.groupby('year')['selling_price'].quantile(0.25)
Q3 = df.groupby('year')['selling_price'].quantile(0.75)
IQR = Q3 - Q1
df['upper_bound'] = df['year'].map(Q3 + 1.5 * IQR)

premium_bikes = df[df['selling_price'] > df['upper_bound']][['name', 'year', 'selling_price', 'yearly_avg_price']]
premium_bikes['premium_pct'] = ((premium_bikes['selling_price'] - premium_bikes['yearly_avg_price']) / premium_bikes['yearly_avg_price']) * 100
```

---

  - Compare with brand-new models (e.g., a ₹190K used Classic 350 vs. ₹210K new).  

- **For Sellers**:  
  - Highlight unique selling points (e.g., "Garage-kept", "First owner").

#3 Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?

   # Relationship Between Track Popularity and Duration

I'll explore whether there's any relationship between a track's popularity and its duration using a scatter plot.

## Analysis

First, let's look at some key statistics:
- **Duration range**: From 89,741 ms (~1:30) to 501,648 ms (~8:22)
- **Average duration**: ~210,000 ms (~3:30)
- **Median duration**: ~198,000 ms (~3:18)

## Visualization

Here's the scatter plot to examine the relationship:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('spotify.csv')

# Convert duration from milliseconds to minutes for better readability
df['Duration (min)'] = df['Duration (ms)'] / 60000

# Create scatter plot
plt.figure(figsize=(12, 7))
plt.scatter(df['Duration (min)'], df['Popularity'], alpha=0.6, color='#1DB954', edgecolor='black')

# Add regression line to show trend
import numpy as np
z = np.polyfit(df['Duration (min)'], df['Popularity'], 1)
p = np.poly1d(z)
plt.plot(df['Duration (min)'], p(df['Duration (min)']), "r--")

plt.title('Relationship Between Track Duration and Popularity', fontsize=14)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('Popularity Score (0-100)', fontsize=12)
plt.grid(alpha=0.3)

# Annotate some outliers
outliers = df[(df['Duration (min)'] > 7.5) | (df['Popularity'] < 40)]
for i, row in outliers.iterrows():
    plt.annotate(f"{row['Artist']} - {row['Track Name']}",
                 (row['Duration (min)'], row['Popularity']),
                 textcoords="offset points", xytext=(0,5), ha='center', fontsize=8)

plt.show()
```

## Key Observations

1. **General Trend**: The red dashed regression line shows a very slight negative trend, suggesting that shorter tracks tend to have marginally higher popularity scores on average.

2. **Concentration**: Most tracks cluster between:
   - 2-4 minutes in duration
   - 70-85 in popularity

3. **Outliers**:
   - Very long tracks (>6 minutes) tend to have lower popularity
   - The longest track (8:22) is DJ Khaled's "GOD DID" with 64 popularity
   - Some very short tracks (<2 minutes) achieve high popularity

4. **Optimal Duration**: The highest concentration of popular tracks (80+ popularity) appears between 2-3.5 minutes.

## Interpretation

While there's a slight tendency for shorter tracks to be more popular, the relationship is weak. Other factors (artist fame, genre, release timing) likely play larger roles in determining popularity. The music industry's general preference for radio-friendly track lengths (around 3 minutes) is reflected in the data clustering.

#8 Visualize relationships between multiple numerical variables simultaneously using a pair plot?

   # **Pair Plot Analysis of Numerical Variables**  

To visualize relationships between **popularity, duration, and other numerical variables**, we'll use a **pair plot** (scatterplot matrix). This helps identify correlations and patterns across multiple dimensions.

### **Key Variables Analyzed**:
1. **Popularity** (0-100)  
2. **Duration (ms)** → Converted to **minutes** for readability  
3. **Track Length (min)** → Derived from duration  

### **Step 1: Data Preparation**
```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('spotify.csv')

# Convert duration from ms to minutes
df['Duration (min)'] = df['Duration (ms)'] / 60000  

# Select numerical columns for pair plot
num_vars = ['Popularity', 'Duration (min)']
```

### **Step 2: Generate Pair Plot**
```python
# Customize plot style
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Create pair plot
pair_plot = sns.pairplot(
    data=df[num_vars],
    kind='scatter',
    diag_kind='kde',  # Kernel Density Estimate for diagonal plots
    plot_kws={'alpha': 0.6, 's': 30},
    height=3
)

# Add title
plt.suptitle("Relationships Between Popularity & Duration", y=1.02, fontsize=14)

plt.show()
```

### **Key Insights from the Pair Plot**:
1. **Popularity vs. Duration (Scatter Plot)**  
   - Most tracks cluster between **2-4 minutes** with **popularity 70-90**.  
   - **No strong correlation** (no clear upward/downward trend).  
   - Extremely long tracks (>6 mins) tend to have **lower popularity**.  

2. **Distribution of Popularity (Diagonal KDE Plot)**  
   - Peaks around **80-85**, indicating most tracks are moderately popular.  
   - Few tracks have very low (<50) or very high (>95) popularity.  

3. **Distribution of Duration (Diagonal KDE Plot)**  
   - Strong peak at **~3 minutes**, the typical length for radio-friendly songs.  

---

### **Enhanced Pair Plot (Optional)**
To add more depth, we can:  
- **Color points by artist** (e.g., highlight Drake vs. Travis Scott).  
- **Add regression lines** to quantify relationships.  

```python
# Highlight top artists
top_artists = ['Drake', 'Travis Scott', 'Post Malone']
df['Highlight'] = df['Artist'].apply(lambda x: x if x in top_artists else 'Other')

# Enhanced pair plot
sns.pairplot(
    data=df,
    vars=num_vars,
    hue='Highlight',
    palette={'Drake': '#1DB954', 'Travis Scott': '#FF6B6B', 'Post Malone': '#4CC9F0', 'Other': '#D3D3D3'},
    plot_kws={'alpha': 0.7, 's': 25}
)
plt.show()
```

### **Conclusion**:
- **No dramatic trends**, but shorter tracks (~3 mins) dominate high popularity.  
- **Drake/Travis Scott** tracks (green/red) are consistently high in popularity.

#9  Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot?

   # **Track Duration Distribution Across Artists**

Let's visualize how track durations vary across different artists using both **box plots** and **violin plots**. We'll focus on the top 10 artists with the most tracks in the dataset for clarity.

## **Analysis & Visualization**

### **1. Data Preparation**
```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('spotify.csv')

# Convert duration from milliseconds to minutes
df['Duration (min)'] = df['Duration (ms)'] / 60000  

# Get top 10 artists by track count
top_artists = df['Artist'].value_counts().head(10).index
df_top = df[df['Artist'].isin(top_artists)]
```

### **2. Box Plot (Duration Distribution by Artist)**
```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_top,
    x='Artist',
    y='Duration (min)',
    palette='viridis',
    order=top_artists  # Sort by track count
)

plt.title("Track Duration Distribution by Artist (Box Plot)", fontsize=14)
plt.xlabel("Artist", fontsize=12)
plt.ylabel("Duration (minutes)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

#### **Key Observations (Box Plot)**:
- **Drake** has a **wide range** (2.5–5.5 mins), with some longer tracks.
- **Post Malone** tends to keep songs **shorter (~2.5–3.5 mins)**.
- **Travis Scott** has a **tight cluster around 3–4 mins**.
- **Lil Wayne** has **more variability**, including some very short and long tracks.

---

### **3. Violin Plot (Density + Distribution)**
```python
plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df_top,
    x='Artist',
    y='Duration (min)',
    palette='magma',
    order=top_artists
)

plt.title("Track Duration Distribution by Artist (Violin Plot)", fontsize=14)
plt.xlabel("Artist", fontsize=12)
plt.ylabel("Duration (minutes)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

#### **Key Observations (Violin Plot)**:
- **Drake** has a **bimodal distribution**, meaning he produces both short and long tracks.
- **Post Malone** has a **peak around 3 mins**, typical for pop/radio-friendly songs.

#10 How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.?

   To visualize the distribution of track popularity for different artists, I'll create a violin plot, which is excellent for showing the distribution and density of data across categories. Here's the approach:

1. Filter the dataset to include only artists with a sufficient number of tracks (e.g., at least 5 tracks) to make meaningful comparisons.
2. Create a violin plot to show the distribution of popularity scores for each artist.

### Violin Plot of Track Popularity by Artist

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('spotify.csv')

# Filter artists with at least 5 tracks
artist_counts = df['Artist'].value_counts()
selected_artists = artist_counts[artist_counts >= 5].index
filtered_df = df[df['Artist'].isin(selected_artists)]

# Create the violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(
    data=filtered_df,
    x='Artist',
    y='Popularity',
    palette='viridis',
    inner='quartile'  # Show quartiles inside the plot
)

# Customize the plot
plt.title('Distribution of Track Popularity by Artist', fontsize=14)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

### Key Observations:
1. **Drake** has consistently high popularity scores, with most tracks above 85, indicating his strong and stable performance on Spotify.
2. **Post Malone** also shows high popularity, but with slightly more variability than Drake.
3. **Nicki Minaj** has a wide range of popularity scores, from the 60s to the high 80s, reflecting her diverse audience reception.
4. **Travis Scott** and **21 Savage** have relatively tight distributions around high popularity scores, suggesting consistent appeal.
5. **Lil Wayne** and **JAY-Z** show broader distributions, with some tracks performing exceptionally well and others moderately.

### Alternative: Swarm Plot
If you prefer a swarm plot (which shows individual data points), you can use:
```python
plt.figure(figsize=(12, 8))
sns.swarmplot(data=filtered_df, x='Artist', y='Popularity', color='black', size=3)
plt.title('Track Popularity by Artist (Swarm Plot)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
